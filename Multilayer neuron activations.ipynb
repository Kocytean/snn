```python
from dsrm import *
import matplotlib.pyplot as plt

# Instantiate a network with 3 layers of 100, 10 and 1 units each
net = Network([100,10,1])

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize = (12,6))
x1, y1, x2, y2, x3, y3 = [], [], [], [], [], []
for i in range(100):
    
    # run the network with some input to input layers
    # you can pass an array of length len(net.layers[0].num_units)
    # or pass a constant input to all units
    
	net.run(0.067)
	for ii, j in enumerate(net.layers[0].state):
		if j:
			x1.append(i)
			y1.append(ii)
	for ii, j in enumerate(net.layers[1].state):
		if j:
			x2.append(i)
			y2.append(ii)
	x3.append(i)
	y3.append(net.layers[0].eta[0] + net.layers[0].epsilon[0])
	Clock.tick()
ax1.scatter(x1, y1, s=2)
ax2.scatter(x2, y2, s=2)
ax3.plot(x3, y3)
fig.show()
```
